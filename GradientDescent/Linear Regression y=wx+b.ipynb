# Backpropagation

$ \large
\begin{align} 
\frac{\partial{J}}{\partial{w}} &= \frac{\partial{}}{\partial{w}}\sum_{i=0}^{N}(y_i-wx_i)^2 \\ 
\large &= 2\sum_{i=0}^{N}(y_i-wx_i)(-x_i)
\end{align}
$

```python
def mse(y_true, y_pred):
    delta = y_true - y_pred
    return (delta**2).mean()
```

```python
w_history = []
b_history = []

loss_history = []
y_pred_history = []

lr = 1e-2

for i in range(50):
    try:
        w.grad.zero_()
        b.grad.zero_()
    except:
        print("Starting the first point")
    w_history.append(w.data.clone())   
    b_history.append(b.data.clone())
    
    # Calculate Loss
    y_pred = x*w + b
    y_pred_history.append(y_pred.data.clone())
    loss = mse(y_true, y_pred)
    loss_history.append(loss.data)
    
    # Calculate gradients
    loss.backward()
    
    # Update parameters
    with torch.no_grad():
        w -= w.grad * lr
        b -= b.grad * lr
```
