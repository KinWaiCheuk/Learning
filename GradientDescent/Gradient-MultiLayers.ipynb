```python
import torch
import matplotlib.pyplot as plt
```

```python
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
def layer1(x):
    return x**2
def layer2(x):
    return x**2
```

```python
x = torch.arange(-10,10.,0.1, requires_grad=True, device=device)
```

```python
y1 = layer1(x)
y2 = layer2(y1)
```

```python
plt.plot(x.cpu().detach().numpy(),y2.cpu().detach().numpy())
```

# autograd.grad approach

```python
dydx, = torch.autograd.grad(y2.sum(),x)
```

```python
plt.plot(x.cpu().detach().numpy(), dydx.cpu().detach().numpy())
```

# Backward Approach

```python
'''The backward will also store the grad in each of the variable'''
```

```python
y2.sum().backward()
```

```python
plt.plot(x.cpu().detach().numpy(),x.grad.cpu().detach().numpy())
```
